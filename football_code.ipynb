```python
import pandas as pd
```

```python
df = pd.read_csv('E:/spi_matches.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df1 = df

l = []
for k,v in df1.iterrows():
    
    # Home Team
    sl = []
    for col in v.index:
        if col[-1] != '1' and col[-1] != '2':
            sl.append(v[col])
    for col in v.index:
        if col[-1] == '1':
            sl.append(v[col])
    sl.append('Home Team')
    l.append (sl)
    
    # Away
    sl = []
    for col in v.index:
        if col[-1] != '1' and col[-1] != '2':
            sl.append(v[col])
    for col in v.index:
        if col[-1] == '2':
            sl.append(v[col])
    sl.append('Away Team')
    l.append (sl)
```

```python
# Print example of first row
print(l[0])
```

```python
# Specify new columns and transform it to DataFrame
col = ['date','league_id','league','team','spi','prob','probtie','proj_score','importance','score','xg','nsxg','adj','home_away']
transformed_df = pd.DataFrame(l,columns = col )

# Check head of dataframe
transformed_df
```

```python
# Export to csv
transformed_df.to_csv('E:/trans_football.csv', header = True)
```
